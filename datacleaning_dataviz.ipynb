```python
import pandas as pd

def impute_delivery_date(df):
    """Imputes missing delivery dates based on order date + 10 days."""

    # Convert 'orderDate' to datetime, handling format and errors
    df['orderDate'] = pd.to_datetime(df['orderDate'], format='%d-%m-%Y', errors='coerce')

    # Convert 'deliveryDate' to datetime, handling errors
    df['deliveryDate'] = pd.to_datetime(df['deliveryDate'], errors='coerce')

    # Find rows where 'deliveryDate' is NaT (originally '?')
    missing_delivery_dates = df['deliveryDate'].isnull()

    # Calculate the imputed delivery dates
    df.loc[missing_delivery_dates, 'deliveryDate'] = df.loc[missing_delivery_dates, 'orderDate'] + pd.Timedelta(days=10)

    return df
```

```python
df = impute_delivery_date(df)
```

```python
def calculate_average_delay(df):
    """Calculates the average delivery delay in days."""

    # Convert date columns to datetime, handling errors
    df['orderDate'] = pd.to_datetime(df['orderDate'],format='%d-%m-%Y', errors='coerce')
    df['deliveryDate'] = pd.to_datetime(df['deliveryDate'], errors='coerce')

    # Calculate the delay in days
    df['delay'] = (df['deliveryDate'] - df['orderDate']).dt.days
    # Filter out negative delays (delivery before order)
    valid_delays = df['delay'][df['delay'] >= 0]

    # Calculate the average delay, ignoring NaT values
    average_delay = valid_delays.mean()

    return average_delay
```

```python
average_delay = calculate_average_delay(df)
print(f"Average Delivery Delay: {average_delay} days")
```

```python
import pandas as pd
import re

def categorize_size(size_str):
    """Categorizes product sizes based on given ranges."""

    # Handle non-numeric sizes
    if pd.isna(size_str):
        return None
    if str(size_str).lower() == 'unsized':
        return 'Unsized'
    if str(size_str).lower() == 'm':
        return 'M'
    if str(size_str).lower() == 'l':
        return 'L'
    if str(size_str).lower() == 'xl':
        return 'XL'
    if str(size_str).lower() == 'xxl':
        return 'XXL'
    if str(size_str).lower() == 'xxxl':
        return 'XXXL'
    if str(size_str).lower() == 's':
        return 'S'
    if str(size_str).lower() == 'xs':
        return 'XS'

    # Remove non-numeric characters and convert to integer if possible
    size_str = re.sub(r'[^\d]', '', str(size_str))

    try:
        size = int(size_str)
    except ValueError:
        return None  # Handle cases where conversion to integer fails

    if size < 34:
        return "others"
    elif size < 36:
        return "XS"
    elif 36 <= size < 38:
        return "S"
    elif 38 <= size < 40:
        return "M"
    elif 40 <= size < 42:
        return "L"
    elif 42 <= size < 44:
        return "XL"
    elif 44 <= size < 46:
        return "XXL"
    elif size >= 46 and size < 1000: #size less than 1000
        return "XXXL"
    elif size >= 1000: #size greater than 1000
        return "Outlier"
    else:
        return None

def group_sizes(df, size_column_name, new_column_name='SizeCategory'):
    """Groups sizes in a DataFrame column into categories."""

    df[new_column_name] = df[size_column_name].apply(categorize_size)
    return df





df = group_sizes(df, 'size')
```
